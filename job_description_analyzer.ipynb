```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import re
%matplotlib inline
```

```python
def analyze(desc):
    
    #lowercase input and split it into a list
    split_desc = desc.lower().split(" ")
    
    #remove all special characters from the input list & remove blanks
    cleaned_list = [re.sub(r"[^a-zA-Z0-9]", "", x) for x in split_desc]
    cleaned_list = list(filter(None, cleaned_list))

    #Read Stopwords file & set stopwords & close file
    file = open('test_stopwords.txt', 'r')
    file_read = file.read()
    stopwords = file_read.splitlines()
    file.close()
    
    #Remove Stopwords from input list
    res = [i for i in cleaned_list if i not in stopwords]
        
    #count the number of words in input list after stopwords are removed
    length_of_list = len(cleaned_list)

    #grab the 10 most common words and put into a tuple
    top_ten =  Counter(res).most_common(10)
    
    #convert tuple to dictionary
    dictionary_of_top = dict(top_ten)
    
    #return keys and values for top 10 results with their percentage
    for k, v in dictionary_of_top.items():
        dictionary_of_top[k] = round(((v/length_of_list)*100),2)
        
    #Create the DataFrame
    df = pd.DataFrame.from_dict(dictionary_of_top, orient='index', columns = ['Word % of Occurence'])
        
    #printing output to user    
    output =  "The following is a list of the top 10 values found in the job description and their associated weighted percentages:"
    print ('{0} \n {1}'.format(output, dictionary_of_top))
    df.plot(kind='bar', figsize=(10,5))
    plt.show()

# next steps:
#read an input file?
#have the program add common words to the stopwords.txt file
#remove numbers from common list

```
